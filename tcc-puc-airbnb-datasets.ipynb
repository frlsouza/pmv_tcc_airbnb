```python
import pandas as pd
import numpy as np
```

```python
df_listings = pd.read_csv('listings.csv', index_col='id', engine='python', sep =',', encoding="utf-8")
print('Total de registros carregados:',len(df_listings))
df_listings.head()
```

```python
df_listings_filtered = df_listings[['host_name', 'host_response_time', 'host_response_rate', 'host_is_superhost', 'host_identity_verified', 'neighbourhood', 'zipcode', 'latitude', 'longitude', 'property_type', 'room_type', 'accommodates', 'bathrooms', 'bedrooms', 'beds', 'bed_type', 'amenities', 'price', 'security_deposit', 'cleaning_fee', 'guests_included', 'extra_people', 'minimum_nights', 'maximum_nights', 'number_of_reviews', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value', 'instant_bookable', 'cancellation_policy', 'calculated_host_listings_count', 'reviews_per_month']]
df_listings_filtered.head()
```

```python
df_listings_filtered.describe()
```

```python
df_listings_filtered_sampled = df_listings_filtered.sample(10000)
```

```python
df_listings_filtered_sampled.describe()
```

```python
dados_ausentes_media = ['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value', 'reviews_per_month']
dados_ausentes_minimo = ['bathrooms', 'bedrooms', 'beds']
df_listings_filtered_sampled.fillna(df_listings_filtered_sampled[dados_ausentes_media].mean(), inplace=True)
df_listings_filtered_sampled.fillna(df_listings_filtered_sampled[dados_ausentes_minimo].min(), inplace=True)
df_listings_filtered_sampled.fillna(value='nd', inplace=True)
df_listings_filtered_sampled.describe()
```

```python
df_listings_filtered_sampled['amenities']
```

```python
amenities_list = df_listings_filtered_sampled['amenities'].str.strip('}{').str.replace(' ', '_').str.replace('"', '').str.split(',')
```

```python
amenities_list
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
amenities_matriz = mlb.fit_transform(amenities_list)
amenities_matriz
```

```python
amenities_colunas = mlb.classes_
amenities_colunas
```

```python
amenities_one_hot_enc = pd.DataFrame(
        mlb.fit_transform(amenities_list),
        index=amenities_list.index,
        columns=mlb.classes_)
```

```python
amenities_one_hot_enc
```

```python
amenities_one_hot_enc.drop(columns='', inplace=True)
amenities_one_hot_enc
```

```python
amenities_one_hot_enc.to_csv('amenities.csv')
```

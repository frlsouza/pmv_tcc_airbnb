```python
import pandas as pd
import numpy as np
```

```python
df_anuncios = pd.read_csv('anuncios.csv', index_col='id', engine='python', sep =',', encoding="utf-8")
df_anuncios.head()
```

```python
df_amenities = pd.read_csv('amenities.csv', index_col='id', engine='python', sep =',', encoding="utf-8")
df_amenities.head()
```

```python
df_metro_quadrado = pd.read_csv('metro_quadrado.csv', index_col='neighbourhood', engine='python', sep =',', encoding="utf-8")
df_metro_quadrado.head()
```

```python
df_anuncios.describe()
```

```python
df_amenities.describe()
```

```python
df_metro_quadrado.describe()
```

```python
df_airbnb_cat = pd.concat([df_anuncios, df_amenities], axis=1)
df_airbnb_cat 
```

```python
df_airbnb_cat.drop(columns='amenities', inplace=True)
df_airbnb_cat
```

```python
df_airbnb_cat = df_airbnb_cat.join(df_metro_quadrado, on='neighbourhood', how='left')
```

```python
df_airbnb_cat.to_csv('airbnb_cat.csv')
```

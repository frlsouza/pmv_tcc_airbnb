```python
import pandas as pd
import numpy as np
```

```python
df_anuncios = pd.read_csv('anuncios.csv', index_col='id', engine='python', sep =',', encoding="utf-8")
df_anuncios.head()
```

```python
df_amenities = pd.read_csv('amenities.csv', index_col='id', engine='python', sep =',', encoding="utf-8")
df_amenities.head()
```

```python
df_metro_quadrado = pd.read_csv('metro_quadrado.csv', index_col='neighbourhood', engine='python', sep =',', encoding="utf-8")
df_metro_quadrado.head()
```

```python
df_airbnb = pd.concat([df_anuncios, df_amenities], axis=1)
df_airbnb 
```

```python
df_airbnb.drop(columns='amenities', inplace=True)
df_airbnb
```

```python
df_airbnb.join(df_metro_quadrado, on='neighbourhood', how='left')
```

```python
# Converte os atributos categóricos em quantitativos discretos.
for column in df_airbnb.columns:
    if str(df_airbnb[column].dtype) not in ['float64', 'int64']:
        df_airbnb[column] = df_airbnb[column].map \
        (dict(zip(df_airbnb[column].unique().tolist(),range(len(df_airbnb[column].unique().tolist())))))
df_airbnb
```

```python
df_airbnb.to_csv('airbnb.csv')
```
